### Goal 1: Group the records by Property ID

```python
# Goal 1: Create a new record for each unique property
records_per_id = la_housing.groupby(['Property ID']).size()
records_per_id.sort_values(ascending=False)
```

```python
# get other aggregation data
apn = la_housing.groupby(['Property ID'])['APN'].apply(list)
category = la_housing.groupby(['Property ID'])['General Category'].apply(list)
subtype = la_housing.groupby(['Property ID'])['Permit Sub-Type'].apply(list)
address = la_housing.groupby(['Property ID'])['Address Full'].apply(list)
units = la_housing.groupby(['Property ID'])['Unit Count'].apply(list)
```

```python
# total number of records per property id
num_records = la_housing.groupby(['Property ID']).size()

# whether property id ever has an ellis withdrawal
has_ellis_withdrawal = la_housing.groupby(['Property ID'])['General Category'].apply(lambda x: "Ellis Withdrawal" in list(x)) 
```

```python
# how to find first index of a value in a series
la_housing['Status Date'] = la_housing['Status Date'].apply(pd.to_datetime)
min_ellis_date = la_housing[la_housing['General Category'] == 'Ellis Withdrawal'].groupby(['Property ID'])['Status Date'].min()
min_ellis_date
```

```python
merged_la_housing = pd.merge(la_housing, pd.DataFrame(min_ellis_date).reset_index(), how="left", left_on= 'Property ID', right_on='Property ID', suffixes=('','_min'))
merged_la_housing.head()
```

```python
merged_la_housing['Before First Ellis'] = merged_la_housing['Status Date'] < merged_la_housing['Status Date_min'] 

merged_la_housing['Entitlement Before First Ellis'] = (merged_la_housing['Status Date'] < merged_la_housing['Status Date_min']) & (merged_la_housing['General Category'] =='Entitlement Change')
```

```python
merged_la_housing.head()
```

```python
grouped_la_housing = pd.DataFrame(la_housing.groupby(['Property ID']).size())

grouped_la_housing.columns = ['Total Actions']

grouped_la_housing
```

```python
grouped_la_housing['Total Actions Before Ellis'] = merged_la_housing.groupby(['Property ID'])['Before First Ellis'].sum()

grouped_la_housing['Total Entitlements Before Ellis'] = merged_la_housing.groupby(['Property ID'])['Entitlement Before First Ellis'].sum()

#grouped_la_housing.to_csv('la_housing_dataset_no_geo_property_id_aggregations_v559.csv')
grouped_la_housing
```
